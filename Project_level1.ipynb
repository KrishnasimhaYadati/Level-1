```python
import socket
import sys
host = 'gaia.cs.umass.edu'
port = 80
sock = socket.create_connection((host, port))
req = ('GET /wiresharklabs/alice.txt HTTP/1.1\r\n'
'Host: {host}:{port}\r\n'
'User-Agent: Python {version}\r\n'
'Connection: close\r\n'
'\r\n')
req = req.format(host=host, port=port, version=sys.version_info[0])
sock.sendall(req.encode('ascii'))
rfc_raw = bytearray()
while True:
    buf = sock.recv(4096)
    if not len(buf):
        break
    rfc_raw += buf
rfc = rfc_raw.decode()
print(rfc)
```

```python
from urllib.request import Request
from urllib.request import urlopen
req = Request('http://www.debian.org')
req.add_header('Accept-Language', 'sv')
print('headers we added:',req.header_items())
response=urlopen(req)
print('server\'s response ',response.readlines()[:50])
print('\nHeaders urlopen() method added',req.header_items)
```

```python
req=Request('http://www.debian.org')
req.add_header("Accept-encoding",'gzip')
response=urlopen(req)
print(f'value{response.getheader("Content-Encoding")}')
print(f'Content: ',response.read())
```

```python
import gzip
content=gzip.decompress(response.read())
print(f"decompressed content:{content.splitlines()[:5]}")
```
